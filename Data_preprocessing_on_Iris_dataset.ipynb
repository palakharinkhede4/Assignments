```python
# Import libraries
import pandas as pd  # For data manipulation
import numpy as np  # For numerical operations
from sklearn.preprocessing import StandardScaler  # For standardization
from sklearn.datasets import load_iris  # For loading the iris dataset
import matplotlib.pyplot as plt  # For data visualization
import seaborn as sns  # For enhanced data visualization

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Create a pandas DataFrame for easier data manipulation
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                  columns=iris['feature_names'] + ['target'])

# Data Cleaning (example: remove rows with missing values)
df.dropna(inplace=True)

# Data Preprocessing (example: select specific features)
df = df[['sepal length (cm)', 'sepal width (cm)', 'target']]

# Standardization
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['sepal length (cm)', 'sepal width (cm)']])
df[['sepal length (cm)', 'sepal width (cm)']] = scaled_data

# Data Visualization
sns.pairplot(df, hue='target')
plt.show()
```
